```python
import numpy as np
import pandas as pd
```

```python
data = pd.DataFrame(list(range(0,100)))
```

```python
X_train = []
for i in range(10, 100):
    X_train.append(data.iloc[i-10:i, 0])
X_train = np.array(X_train)
pd.DataFrame(X_train)
```

```python
reshape = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
reshape
```
